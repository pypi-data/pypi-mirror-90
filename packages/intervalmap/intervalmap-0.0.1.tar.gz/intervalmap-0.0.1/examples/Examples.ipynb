```python
import intervalmap as intmap
```

```python
def f(x):
    return x**5 - 8 * x**3 + 10 * x + 6
```

```python
x_range = [-2.7, 2.7]
num_pts = 100

intmap.PlotMap(f, x_range, num_pts)
```

```python
x_range = [-1, 1]
num_pts = 20

intmap.PlotIntervalMap(f, x_range, num_pts)
```

```python
x_range = [-1, 1]
num_pts = 50

intmap.PlotIntervalMap(f, x_range, num_pts)
```
