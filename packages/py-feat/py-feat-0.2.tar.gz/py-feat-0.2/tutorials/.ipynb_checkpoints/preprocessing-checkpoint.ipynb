# This is a typical preprocessing pipeline for analyzing facial expression data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches

import numpy as np

from feat.data import Facet, Fex
from feat.utils import read_facet
```

```python
# Load sample data
fname = '../feat/tests/data/iMotions_Test_v6_long.txt'
dat = Fex(read_facet(fname))
```

# Rectification

```python
"""
Plot the FaceRectX, and FaceRectY coordinates.
Outliers from these coordinates indicate that the face has moved substantially from the average position,
often due to either face detection failure or occlusion of the face. 
This test data was filmed with a head mounted camera so we know this is due to face occlusion. 
"""
f,ax = plt.subplots(figsize=(15,4))
dat['FaceRectX'].plot()
dat['FaceRectY'].plot()
start, end,  = dat.index.min(), dat.index.max()
stepsize = 5000
plt.xticks(np.arange(start, end, stepsize),rotation='vertical')
plt.show()
```

We can see that there is an outlier in head position between 90,000 and 95,000 timepoint
Let's take a closer look at those timepoints with an overlay if predictions our NaNs 

```python
f,ax = plt.subplots(figsize=(15,2))
dat['FaceRectX'].plot()
start, end,  = dat.index.min(), dat.index.max()
stepsize = 10000
ax.xaxis.set_ticks(np.arange(start, end, stepsize))
[ax.axvline(loc, linewidth=1, color='r', alpha=.05) for loc in np.where(dat['Joy'].isna())[0]]
ax.set(xlim=[90000,95000])
plt.show()
```

We can see that there is an outlier in head position between 90,000 and 10,000 timepoint. 
We can also plot the rectangle of the face as below to verify how much it is moving

```python
def plot_facerect(_dat):
    """ Plot FaceRect to show how much face is shifting. 
    """
    f, ax = plt.subplots()
    for rowix, row in _dat.iterrows():
        rect = patches.Rectangle((row.FaceRectX,row.FaceRectY),
                                 row.FaceRectWidth,row.FaceRectHeight,linewidth=1,edgecolor='r',facecolor='none')
        ax.add_patch(rect)
    ax.set(xlim=[0,1280],ylim=[0,720])
    plt.gca().invert_yaxis()
    plt.show()
plot_facerect(dat.iloc[90000:95000])
```

We can use the `rectification` function to remove these outliers based on some standard deviation from the average position. 
By default it will rectify based on both X and Y position. 

```python
# Remove FaceRect that diverge more than 3 standard deviations from average 
std = 3
rectified_dat = dat.rectification(std=std)
f,ax = plt.subplots(2,1,figsize=(15,4),sharey=True)
for col in ['FaceRectX','FaceRectY']:
    dat[col].plot(ax=ax[0],label=col)
    rectified_dat[col].plot(ax=ax[1],label=col)
ax[0].set(title="Before Rectification")
ax[1].set(title="After Rectification")
start, end, stepsize = dat.index.min(), dat.index.max(), 10000
ax[0].xaxis.set_ticks(np.arange(start, end, stepsize))
ax[1].xaxis.set_ticks(np.arange(start, end, stepsize))
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plot_facerect(rectified_dat.iloc[90000:95000])
```

# Interpolation

```python
f,ax = plt.subplots(figsize=(15,4))
dat.loc[90000:95000,'Joy'].plot()
[ax.axvline(loc, linewidth=.5, color='r', alpha=.2) for loc in np.where(dat['Joy'].isna())[0]]
plt.xlim([91000,92000])
plt.show()
```

```python
f,ax = plt.subplots(figsize=(15,4))
dat.loc[90000:95000,'Joy'].plot()
interpolated = dat.interpolate(limit=120)
nulls = interpolated.loc[np.where(dat.Joy.isna())[0],'Joy']
plt.scatter(nulls.index,nulls.values,color='r',s=5)
plt.xlim([91000,92000])
```
